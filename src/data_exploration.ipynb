```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#loading the processed dataframe
df4 = pd.read_csv('../data/processed_data_train.csv')
```

```python
#ploting to check surplus values
df_surplus = df4.filter(regex='StartTime|surplus')
df_total_surplus = pd.DataFrame({'Total_surplus': df_surplus.sum(axis=1)})

plt.figure(figsize=(20, 2))
plt.hist(df_total_surplus['Total_surplus'], bins=50, color='blue', alpha=0.7)
plt.title('Surplus Distribution')
plt.xlabel('Surplus')
plt.ylabel('Frequency')
plt.show()
```

```python
# Assuming df_surplus contains the relevant data

# Set the style of seaborn
sns.set(style="whitegrid")

# Create boxplots for each column
plt.figure(figsize=(20, 3))
sns.boxplot(data=df_surplus, palette="Set3")

# Customize the plot
plt.title('Distribution of Surplus Values')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.show()
```

```python
# Set the style of seaborn
sns.set(style="whitegrid")

# Calculate the total for each column
df_surplus['Total'] = df_surplus.sum(axis=1)

# Assuming df_surplus['StartTime'] is in datetime format
plt.figure(figsize=(20, 3))
sns.lineplot(x='StartTime', y='Total', data=df_surplus)

# Calculate the number of ticks to display
num_ticks = 10  # Adjust this based on your preference
tick_positions = range(0, len(df_surplus), len(df_surplus) // num_ticks)

# Customize the plot
plt.title('Total Surplus Over Time')
plt.xticks(tick_positions, df_surplus['StartTime'].iloc[tick_positions], rotation=45, ha='right')
plt.show()
```

```python
# Plotting the countplot
plt.figure(figsize=(10, 2))
sns.countplot(x='y', data=df4, palette='viridis')
plt.title('Frequency of Values in Column y')
plt.xlabel('y')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting the line plot
plt.figure(figsize=(10, 4))
plt.plot(df4.index, df4['y'], marker='o', linestyle='-', color='b', linewidth=0.25)
plt.title('Evolution of Column y Through Time')
plt.xlabel('Time')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
